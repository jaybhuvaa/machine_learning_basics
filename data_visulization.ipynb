```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
```

```python
details=pd.read_csv('details.csv')
placement=pd.read_csv('placement.csv')
details=pd.DataFrame(details)
palcement=pd.DataFrame(placement)
```

```python
details
```

```python
placement
```

```python
merge=pd.merge(details,palcement,on='Roll No.',how='outer')
merge
```

```python
placed=merge[merge['Placed']=='yes']
placed
```

```python
placed.reset_index(inplace=True,drop=True)
placed
```

```python
salary=[]

for i in range(28):
    cpi=placed.loc[i].at['CPI']
    if(cpi>=8.5):
        sal=random.randint(65000,80000)
        salary.append(sal)
    elif(cpi>=7.5):
        sal=random.randint(50000,65000)
        salary.append(sal)
    elif(cpi>=6.5):
        sal=random.randint(35000,50000)
        salary.append(sal)
    else:
        sal=random.randint(20000,35000)
        salary.append(sal)
        
placed.insert(17,'Salary',salary,True)
placed
        
```

```python
placed.to_csv('Placed.csv',index=True)
```

```python
min_salary = placed['Salary'].min()
max_salary = placed['Salary'].max()

for i in range(28):
    salary = placed.loc[i].at['Salary']
    num = salary - min_salary
    den = max_salary - min_salary
    salary1 = num/den
    placed.at[i,'Salary']=salary1
placed
```

```python
placed.describe()
```

```python
cpi_col = placed['CPI']
salary_col = placed['Salary']

plt.plot(cpi_col,salary_col,marker='o')
plt.xlabel('cpi')
plt.ylabel('salary')
plt.title('cpi vs salary')
```
