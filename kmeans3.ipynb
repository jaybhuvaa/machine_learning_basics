```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
income = pd.read_csv("income.csv")
income = pd.DataFrame(income)
income
```

```python
x = income['Age']
y = income['Income']
```

```python
plt.scatter(x,y)
plt.xlabel("age")
plt.ylabel("income")
```

```python
kmeans = KMeans(n_clusters=3)
kmeans
```

```python
cluster_labels = kmeans.fit_predict(income[['Age','Income']])
income['Cluster']=cluster_labels
income
```

```python
clust1 = income[income.Cluster==0]
clust2 = income[income.Cluster==1]
clust3 = income[income.Cluster==2]
```

```python
clust1
```

```python
clust2
```

```python
clust3
```

```python
plt.scatter(clust1['Age'],clust1['Income'],color='red')
plt.scatter(clust2['Age'],clust2['Income'],color='blue')
plt.scatter(clust3['Age'],clust3['Income'],color='green')
```

```python
# normalising
max_age = income['Age'].max()
max_income = income['Income'].max()
print(max_age)
print(max_income)
```

```python
income['Age'] = income['Age']/max_age
income['Income'] = income['Income']/max_income
```

```python
income
```

```python
kmeans = KMeans(n_clusters=3)
kmeans
```
