```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import MinMaxScaler
```

```python
data=pd.read_csv('most_runs_average_strikerate.csv')
data=pd.DataFrame(data)
data
```

```python
x=data['total_runs']
y=data['out']
```

```python
plt.scatter(x,y)
plt.xlabel('total_runs')
plt.ylabel('out')
```

```python
kmeans=KMeans(n_clusters=3)
kmeans
```

```python
cluster_labels=kmeans.fit_predict(data[['total_runs','out']])
data['cluster']=cluster_labels
data
```

```python
c1=data[data.cluster==0]
c2=data[data.cluster==1]
c3=data[data.cluster==2]
```

```python
max_run=data['total_runs'].max()
print(max_run)
max_out=data['out'].max()
print(max_out)
```

```python
data['total_runs']=data['total_runs']/max_run
data['out']=data['out']/max_out
data
```
